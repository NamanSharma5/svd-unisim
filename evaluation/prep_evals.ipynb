```python
image_path_to_label = {}
image_path_prefix = "/vol/biomedic3/bglocker/ugproj2324/nns20/svd-unisim/"
with open("/vol/biomedic3/bglocker/ugproj2324/nns20/svd-unisim/evaluation/in_distribution_actions_extended.txt", "r") as f:
    lines = f.readlines()
    for line in lines:
        image_path = line.split(",")[0].strip()
        label = line.split(",")[-1].strip()
        image_path = image_path_prefix + image_path
        print(image_path)
        image_path_to_label[image_path] = label
```

```python
from IPython.display import Image

# Assuming you have a dictionary 'image_path_to_label' mapping image paths to labels
for image_path, label in image_path_to_label.items():
    print(f"Label: {label}")
    print(f"Image path: {image_path}")
    Image(filename=image_path)
    display(Image(filename=image_path))
```
